### 2. String cleaner
Given the string `" Adenosine Triphosphate "`, do the following:  
1. Strip spaces  
2. Convert to lowercase  
3. Replace "a" with "@"

```python
import re
def stripSpaces(s):
    return s.strip()

def lowerCase(s):
    return s.lower()

def replace(s):
    return re.sub(r'[aA]', '@', s)

s = " Adenosine Triphosphate "
print(stripSpaces(s))
print(lowerCase(s))
print(replace(s))
    
```

Print numbers from 1 to 30, but:

### 4. Guess the number
The program generates a random number between 1 and 20.  
The user must guess until correct.  
Print hints "too high" or "too low".

```python
import random

randomNum = random.randint(1,21)
print(randomNum)
guess = int(input("Guess the number between 1 and 20: "))
while(guess != randomNum):
    if(guess < randomNum):
        print("Too low")
    else:
        print("Too high")
    guess = int(input("Guess the number between 1 and 20: "))
print("Correct!")
```

### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial).

Provide two solutions:
A) One using a for loop.
B) One using a while loop. (Do not use recursion)

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
# Solution A: Using for loop
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"{num}! = {factorial}")

# Solution B: Using while loop
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"{num}! = {factorial}")
```

### 1. Palindrome
Write `is_palindrome(word)` returning `True` if `word` reads the same backwards.  

```python
import re
def checkPalindrome(s):
    s.lower()
    if s == s[::-1]:
        return True
    else:
        return False

s = input("Enter a string: ")
checkPalindrome(s)
    
```

### 2. Normalize
Write a function `normalize(xs)` that takes a list of numbers and returns a new list scaled to sum to 1.
- If the sum is 0, return a list of zeros of the same length.

```python
def normalize(xs):
    for i in range(len(xs)):
        xs[i] = i + 1
    if xs[i] == 0:
        xs[i] = 1
    return xs
        

xs = [0]
xs2 = [54, 23, 1, 12]
normalize(xs), normalize(xs2)
```

### 3. Frequency table
Implement `freq(s)` returning a dict `{char: count}` for a string.

```python
def freq(s):
    d = {}
    for ch in s:
        if ch in d:
            d[ch] += 1
        else:
            d[ch] = 1
    print(d)
    if d:
        max_count = max(d.values())
    else:
        max_count = 0
    print(max_count)

s = "hello world"
freq(s)
        
```

### 4. — List/dict comprehension
Given `records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]`, build:
1) a dict `grades`
2) a list of names with grade >= 14

```python
records = [('Ana', 15), ('Bruno', 20), ('Carla', 17)]
grades = {}
for names in records:
    if names[1] >= 14:
        grades[names[0]] = names[1]
print(grades)
```

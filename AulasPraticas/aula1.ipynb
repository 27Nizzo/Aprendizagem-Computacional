### 1. Age checker
Ask for the user’s age and print:
- "Minor" if < 18  
- "Adult" if between 18–65  
- "Senior" if > 65

```python
age = int(input("Your age: "))
if(age < 18):
    print("Minor")
elif (age <= 65):
    print("Adult")
else:
    print("Senior")
```

### 2. String cleaner
Given the string `" Adenosine Triphosphate "`, do the following:  
1. Strip spaces  
2. Convert to lowercase  
3. Replace "a" with "@"

```python
import re
def stripSpaces(s):
    return s.strip()

def lowerCase(s):
    return s.lower()

def replace(s):
    return re.sub(r'[aA]', '@', s)

s = " Adenosine Triphosphate "
print(stripSpaces(s))
print(lowerCase(s))
print(replace(s))
    
```

### 3. CompScience FizzBuzz
Print numbers from 1 to 30, but:  
- Replace multiples of 3 with "Comp"  
- Replace multiples of 5 with "Science"  
- Replace multiples of both with "CompScience"

```python
for i in range(1,31):
    if(i % 3 == 0 and i % 5 == 0):
        print("CompScince")
    elif(i % 3 == 0):
        print("Comp")
    elif(i % 5 == 0):
        print("Science")
    else:
        print(i)
```

### 4. Guess the number
The program generates a random number between 1 and 20.  
The user must guess until correct.  
Print hints "too high" or "too low".

```python
import random

randomNum = random.randint(1,21)
print(randomNum)
guess = int(input("Guess the number between 1 and 20: "))
while(guess != randomNum):
    if(guess < randomNum):
        print("Too low")
    else:
        print("Too high")
    guess = int(input("Guess the number between 1 and 20: "))
print("Correct!")
```

### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial).

Provide two solutions:
A) One using a for loop.
B) One using a while loop. (Do not use recursion)

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
# Solution A: Using for loop
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"{num}! = {factorial}")

# Solution B: Using while loop
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"{num}! = {factorial}")
```

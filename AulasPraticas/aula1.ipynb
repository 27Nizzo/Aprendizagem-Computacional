### 2. String cleaner
Given the string `" Adenosine Triphosphate "`, do the following:  
1. Strip spaces  
2. Convert to lowercase  
3. Replace "a" with "@"

```python
import re
def stripSpaces(s):
    return s.strip()

def lowerCase(s):
    return s.lower()

def replace(s):
    return re.sub(r'[aA]', '@', s)

s = " Adenosine Triphosphate "
print(stripSpaces(s))
print(lowerCase(s))
print(replace(s))
    
```

### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial).

Provide two solutions:
A) One using a for loop.
B) One using a while loop. (Do not use recursion)

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
# Solution A: Using for loop
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"{num}! = {factorial}")

# Solution B: Using while loop
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1
print(f"{num}! = {factorial}")
```

### 2. Normalize
Write a function `normalize(xs)` that takes a list of numbers and returns a new list scaled to sum to 1.
- If the sum is 0, return a list of zeros of the same length.

```python
def normalize(xs):
    for i in range(len(xs)):
        xs[i] = i + 1
    if xs[i] == 0:
        xs[i] = 1
    return xs
        

xs = [0]
xs2 = [54, 23, 1, 12]
normalize(xs), normalize(xs2)
```

### 3. Frequency table
Implement `freq(s)` returning a dict `{char: count}` for a string.

```python
def freq(s):
    d = {}
    for ch in s:
        if ch in d:
            d[ch] += 1
        else:
            d[ch] = 1
    print(d)
    if d:
        max_count = max(d.values())
    else:
        max_count = 0
    print(max_count)

s = "hello world"
freq(s)
        
```

### 4. — List/dict comprehension
Given `records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]`, build:
1) a dict `grades`

### 2. JSON

A) Create a dictionary representing a lab sample with properties like `name`, `pH`, `temperature`, and `date_collected`.

B) Save this dictionary as a JSON file called `sample.json`.

C). Read the JSON file back and print the sample information in a formatted way.

```python
import json 
lab_sample = {
    "name": "sample_1",
    "ph": 7.2,
    "temperature": 25.0,
    "date_collected": "2024-01-15"
}

print(json.dumps(lab_sample, indent=4))

files = [
    ("sample_default.json", {}),
    ("sample_pretty.json", {"indent": len(lab_sample)})
]

for filename, options in files:
    with open(filename, "w") as f: 
        json.dump(lab_sample, f, **options)

# Read the JSON file back and print the sample information
for filename, options in files:
    with open(filename, "r") as f:
        sample_data = json.load(f)
        print(f"\n{filename}:")
        print(json.dumps(sample_data, **options))
```

### 4. Recursion
A) Write a recursive function to calculate the power of a number: `power(base, exponent)`.

$$
2^4 = 2 \times 2 \times 2 \times 2 = 16
$$


B) Write a recursive function that counts down from a given number to 1, printing each number.

```python
#A) 
def powerOfNum(base, exponent):
    if exponent == 0 or base == 1:
        return 1
    else:
        return base * powerOfNum(base, exponent - 1)
    
print(powerOfNum(2, 3))
    
```

```python
#B) 

def countDown(n):
    if n == 0:
        print("Blast off!")
    else:
        print(n)
        countDown(n - 1)
countDown(5)
```

### 5. Classes: Library Book Management

**Problem:**

Create a Python class named `Book` to represent books in a library. Each `Book` object should have the following attributes:

- `title` (string)
- `author` (string)
- `isbn` (string, a unique identifier for the book)
- `is_available` (boolean, `True` if the book is available, `False` otherwise, default to `True`)

The `Book` class should also have the following methods:

- `__init__(self, title, author, isbn)`: Initializes the book with the given title, author, and ISBN. `is_available` should be set to `True` by default.
- `borrow_book(self)`: Changes the `is_available` status to `False` if the book is currently available. It should print a message indicating whether the book was successfully borrowed or if it's already on loan.
- `return_book(self)`: Changes the `is_available` status to `True` if the book is currently on loan. It should print a message indicating whether the book was successfully returned or if it was already available.
- `__str__(self)`: Returns a string representation of the book in the format: `"<Title> by <Author> (ISBN: <ISBN>) - Status: <Available/On Loan>"`

**Task:**

1. Define the `Book` class with the attributes and methods described above.
2. Create at least two `Book` objects.
3. Demonstrate the usage of `borrow_book`, `return_book`, and `__str__` methods by borrowing and returning books and printing their status.

```python
class Book:
    def __init__(self, tittle, author, isbn, is_available):
        self.tittle = tittle
        self.author = author
        self.isbn = isbn
        self.is_available = is_available
        
    def __str__(self):
        return f"Title: {self.tittle},\n Author: {self.author},\n ISBN: {self.isbn},\n Available: {self.is_available}"
    
    def borrow_book(self):
        if self.is_available:
            self.is_available = False
            print(f"The book '{self.tittle}' has been borrowed.")
        else:
            print(f"The book '{self.tittle}' is not available.")
    
    def return_book(self):
        if not self.is_available:
            self.is_available = True
            print(f"The book '{self.tittle}' has been returned.")
        else:
            print(f"The book '{self.tittle}' was not borrowed.")
            
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "978-0743273565", True)
print(book1)
print("--------")
book1.borrow_book()
print(book1)
print("--------")
book1.return_book()
print(book1)
```
